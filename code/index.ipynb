### Introduction:

In recent years, the housing market in the US has undergone significant fluctuations, leading to changes in housing prices across the country. Over the past few decades, the US housing market has undergone significant changes. The 1990s and early 2000s saw an increase in housing prices, driven by low-interest rates, relaxed lending standards, and a robust economy. However, this housing boom eventually led to a housing market crash and financial crisis in 2008, with housing prices plummeting, and foreclosures skyrocketing.

Since the 2008 housing crisis, the US housing market has experienced a slow but steady recovery. Housing prices have been increasing steadily over the past decade, albeit with variations across different regions and states. Despite the recovery, several challenges and uncertainties remain, such as the impact of the COVID-19 pandemic on the housing market, demographic changes, and the affordability of housing for low-income households.

Understanding the factors that drive the US housing market’s trends and dynamics is crucial for investors, and households. For investors, understanding the housing market’s dynamics can help them make informed investment decisions. For households, understanding the housing market can make a proper decision on purchasing.

This research aims to investigate several aspects of the US housing market, including the intrinsic factors that affect housing prices, the state-level variations in housing prices, the trends in housing prices over the past few decades, and the relationship between crime rates, population, and salaries with housing prices.



#### Research questions:

1. What is the trend of the US housing price from 1975 to 2022?
2. Which states have relatively higher housing prices in the US? And which states have relatively lower housing prices?
3. Does the seasonality affect the US housing price?
4. What are the intrinsic factors that contribute to higher housing prices?
5. How does the crime rate affect the US housing price?
6. What specific types of crimes have the most significant impact on housing prices? Are violent crimes more influential than property crimes, or vice versa?
7. Are there any notable exceptions to the general trend of higher crime rates leading to lower housing prices? If so, what factors contribute to these exceptions?
8. Is there any correlation between the housing price and the average salaries of a state?
9. Is there any correlation between the housing price and the population of a state?

Overall, this research aims to provide insights into the various factors that affect the housing price as well as driving the housing market's trends. The results of this research may have important implications for investors and households, enabling them to make more informed decisions related to the housing market.





### Methods:

##### 1. Trend of the Housing Price over Year in Each State (@sec-1)

To analyze the housing price in each state over year, firstly, calculate the average house index from 1975 to 2022 in each state. Then plot the bar plot that shows the mean of the house price indexes in each state from 1975 to 2022 and sorted in descending order based on the value of the house index. Finally, create a line plot that shows the trend of the house price index in each state from 1975 to 2022. By clicking on the bar plot of a selected state, the corresponding line plot for that state will be shown individually.


#### 2. Seasonality of the Housing Price (@sec-2)

The data used to create the graphs were obtained from the Federal Housing Finance Agency's House Price Index Datasets. Quarterly purchase-only indexes of US states were employed to generate the graphs. The bar plots were created using the bar function in plotly.express, with the x-axis representing the states of the United States and the y-axis representing the housing index of each state. Additionally, a slider component was added to adjust the time, and the data of different quarters were represented by different colors. The distinction between the two bar plots is that the former merged the seasonal data, whereas the latter displayed the data separately.

##### 3. Intrinsic Factors that Contribute to a Higher Housing Price (@sec-3)

To deal with the intrinsic factors that contribute to a higher housing price, firstly create multiple data sets including a single factor and the housing price. The average housing price is then calculated for each category within each single factor. Use a box plot to display the distribution of housing prices for each category within each factor, then remove the outliers that fall outside the range of upper quarter+1.5IQR and lower quarter-1.5IQR. Finally, use the "plotly" method to generate an interactive plot with the housing price on the y-axis and the selected factor on the x-axis for each subplot. This graph will effectively demonstrate the relationship between housing prices and the selected factors. A correlation plot between the level of the housing price and those factors is drawn after this graph to help better select the relevant factors.

##### 4. Housing Price with Crime Rate (@sec-4)

The correlation between the housing price index and violent crime rates in the United States was analyzed using Plotly Express and Seaborn visualization techniques. Firstly, a choropleth map was created through Plotly Express to represent the housing price index by state. The map displayed a color-coded representation of higher or lower values, enabling users to access detailed information about cities and crime rates interactively. Next, an interactive scatter plot was generated to explore the relationship between the housing price index and violent crime rate per 100,000 population for each state. The color-coded points depicted the states, and users could extract insightful details by interacting with the plot. Lastly, the Seaborn library was utilized to construct a line plot that showed the relationship between the average housing price index and the number of violent crimes over time. The plot included two separate y-axes that displayed both variables, allowing observation of trends. These visualizations provided valuable insights into the correlation between housing prices and crime rates across different regions and time periods.

#### 5. Relationship between the Housing Price and Salary & Population (@sec-5)

To begin the analysis, three separate datasets were merged, each containing information on housing prices, salaries, and population, respectively. For the housing price dataset, filtered it to only include data in 2022 and created a column with the full names of each state based on the state abbreviation column. For salaries and popualtion datasets, remove the special characters. After merging the datasets, two new columns were created to contained text information about the relationship between housing prices and salaries, as well as the relationship between housing prices and population. Finally, create two map plots using the Scattergeo() function, one for the relationship between housing prices and salaries, and the other for the relationship between housing prices and population. In each map plot, the size of the markers was determined by the salaries or population of the state, respectively, while the color of the markers was based on the housing price index. The range of the color bar was set according to the range of the housing price index. By hovering over a marker on the map, users can see text information about that particular state.

```python
# | echo: false
import numpy as np
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
import plotly.io as pio
import altair as alt
import seaborn as sns
import pandas as pd
import plotly.graph_objs as go
import plotly.offline as pyo
import plotly.subplots as sp
import matplotlib.pyplot as plt
pio.renderers.default = "plotly_mimetype+notebook_connected"
```


#### 1. Trend of the Housing Price over Year in Each State {#sec-1}

```python
# | echo: false
df1=pd.read_csv("../data/final_report/Housing.csv")
df1=df1.drop(columns=['hotwaterheating', 'prefarea'])

df1.rename(columns = {'price':'Price'}, inplace = True)
df1.rename(columns = {'area':'Area'}, inplace = True)
df1.rename(columns = {'bedrooms':'Bedrooms'}, inplace = True)
df1.rename(columns = {'bathrooms':'Bathrooms'}, inplace = True)
df1.rename(columns = {'stories':'Stories'}, inplace = True)
df1.rename(columns = {'mainroad':'Mainroad'}, inplace = True)
df1.rename(columns = {'guestroom':'Guestroom'}, inplace = True)
df1.rename(columns = {'basement':'Basement'}, inplace = True)
df1.rename(columns = {'airconditioning':'AirConditioning'}, inplace = True)
df1.rename(columns = {'parking':'Parking'}, inplace = True)
df1.rename(columns = {'furnishingstatus':'FurnishingStatus'}, inplace = True)

# create data frame with selected object
df1_area=df1[["Price","Area"]].groupby(["Area"]).mean().reset_index()

df1_bed=df1[["Price","Bedrooms"]].groupby(["Bedrooms"]).mean().reset_index()

df1_bath=df1[["Price","Bathrooms"]].groupby(["Bathrooms"]).mean().reset_index()

df1_stories=df1[["Price","Stories"]].groupby(["Stories"]).mean().reset_index()

df1_mainroad=df1[["Price","Mainroad"]].groupby(["Mainroad"]).mean().reset_index()

df1_guestroom=df1[["Price","Guestroom"]].groupby(["Guestroom"]).mean().reset_index()

df1_basement=df1[["Price","Basement"]].groupby(["Basement"]).mean().reset_index()

df1_air=df1[["Price","AirConditioning"]].groupby(["AirConditioning"]).mean().reset_index()

df1_parking=df1[["Price","Parking"]].groupby(["Parking"]).mean().reset_index()

df1_fur=df1[["Price","FurnishingStatus"]].groupby(["FurnishingStatus"]).mean().reset_index()
```

```python
# | echo: false
df_house=df1.copy()

varlist =  ['Mainroad', 'Guestroom', 'Basement', 'AirConditioning']

# Defining the map function
def binary_map(x):
    return x.map({'yes': 1, "no": 0})

# Applying the function to the housing list
df_house[varlist] = df_house[varlist].apply(binary_map)

def fur_map(x):
    return x.map({'furnished': 0, "semi-furnished": 1, 'unfurnished': 2})

df_house[["FurnishingStatus"]]=df_house[["FurnishingStatus"]].apply(fur_map)
```

```python
# | echo: false

df2=pd.read_csv("../data/final_report/HPI_AT_state.csv",header = None)
df2.columns=['State','Year','Quarter','H_idx']

df2_year=df2[["State","Year","H_idx"]].groupby(["State","Year"]).mean().reset_index()
# df2_year
```

```python
# | echo: false
# The bar plot 
# make the bar can be selected, and selected based on the name od the appliances
selection = alt.selection_single(fields=['State'],name='Random')
color = alt.condition(selection,
                      alt.value('steelblue'),         # the selected bar's color is steelblue
                      alt.value('lightgray'))         # unselected bars is lightgray

bar=(alt.Chart(df2_year)
 .mark_bar()
 .encode(y='mean(H_idx):Q',            # y is the value of the sum of the power
         x=alt.X('State:N',            # x is the appliances, and each appliances is the sum of the power, 
         sort=alt.EncodingSortField(field='H_idx', op='mean',     
                            order='descending')),  # and they are in descending order according to the sum of the power
         color=color
    
        )
).add_selection(selection)           # the selection and the color are defined above

# set for the title, xlabel, and ylabel.
bar.title ="Mean of the House Price Indexes in each State in US from 1975 to 2022"
bar.encoding.x.title = 'State'
bar.encoding.y.title = 'Mean of the House Price Indexes'


# The corresponding line plot for the selected appliances.
color2 = alt.condition(selection,
                      alt.Color('State:N', legend=None),                # The color is based on the selected appliance
#                       alt.value('steelblue'),
                      alt.value('lightgray'))             # unselected is dark gray

line1=(alt.Chart(df2_year)
 .mark_line()                                                     # line plot
 .encode(x=alt.X('Year:O',axis=alt.Axis(labelAngle=-45)),          # x is the hour
         y='mean(H_idx):Q',                                      # y is the sum of the power
         color=color2,                                 # selection of the color is defined above
         opacity=alt.condition(selection, alt.value(1.0), alt.value(0.1))
    
        )
)

# set the title, x label, and y label
line1.title ="House Price Indexes in each State in US from 1975 to 2022"
line1.encoding.x.title = 'Year'
line1.encoding.y.title = 'House Price Indexes'

# plot horizontally 
bar & line1
```

<br>

The bar plot indicates that the mean house price index in MA is the highest, which can be attributed to inflation, land costs, labor, and materials in the region. New York and Washington DC also exhibit relatively high mean house price indices due to the fact that DC is the capital of the US and NY serves as the center of the US economy. Conversely, WV has the lowest mean house price index, likely because population centers in the state are relatively small, resulting in a lower tax base for services compared to other regions of the country.

The line plot provides a clear indication that the US housing price index is generally in an upward trend across all states. However, during the period of 2007 to 2012, the housing price index sharply declined for the majority of states due to the financial crisis. Since then, the index has steadily increased and remains on an upward trajectory, with an even higher rate of increase than before the crisis.

## Scatter Plot


```python
# | echo: false
data = pd.read_csv('../data/final_report/crime_rate.csv')

data = data.dropna(subset=['Assaults'])
data[['City', 'State_temp']] = data['City, State'].str.split(', ', expand=True)
data.rename(columns={'State_temp': 'State'}, inplace=True)
```

```python
# | echo: false
# Calculate the violent crime rate per 100,000 population
data['Violent Crime Rate'] = data['Violent Crimes'] / data['Population'] * 100000

# Create an interactive scatter plot using Plotly
fig = px.scatter(
    data,
    x='index_nsa',
    y='Violent Crime Rate',
    color='State',
    hover_data=['City', 'Year', 'Population'],
    labels={
        'index_nsa': 'Housing Price Index',
        'Violent Crime Rate': 'Violent Crime Rate per 100,000 Population'
    },
    title='Housing Price Index vs. Violent Crime Rate',
    width=800,
    height=600
)

fig.show()
```

## Map


```python
# | echo: false
# Load the dataset
data = pd.read_csv('../data/final_report/crime_rate.csv')

data = data.dropna(subset=['Assaults'])
data[['City', 'State_temp']] = data['City, State'].str.split(', ', expand=True)
data.rename(columns={'State_temp': 'State'}, inplace=True)

# Choropleth map with Plotly
fig = px.choropleth(data, locations='State', locationmode='USA-states', color='index_nsa',
                    scope='usa', hover_name='City', hover_data=['Violent Crimes', 'Homicides', 'Rapes', 'Assaults', 'Robberies'],
                    title='Housing Price Index by State')
fig.show()
```

## line Plot


```python
# | echo: false
# Calculate average housing prices and crime rates by year
yearly_data = data.groupby('Year')[['index_nsa', 'Violent Crimes', 'Homicides', 'Rapes', 'Assaults', 'Robberies']].mean().reset_index()

# Create a line plot using Seaborn
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

sns.lineplot(data=yearly_data, x='Year', y='index_nsa', ax=ax1, color='blue')
sns.lineplot(data=yearly_data, x='Year', y='Violent Crimes', ax=ax2, color='red')

ax1.set_xlabel('Year')
ax1.set_ylabel('Housing Price Index', color='blue')
ax2.set_ylabel('Violent Crimes', color='red')

plt.title('Housing Price Index and Violent Crimes Over Time')

# Create a custom legend
from matplotlib.lines import Line2D
custom_lines = [Line2D([0], [0], color='blue', lw=4),
                Line2D([0], [0], color='red', lw=4)]
ax1.legend(custom_lines, ['Housing Price Index', 'Violent Crimes'], loc='upper left', bbox_to_anchor=(0.1, 0.9))

plt.show()
```

The interactive scatter plot further supports this observation, showing a negative correlation between the housing price index and violent crime rate per 100,000 population for each state. However, there are some outliers, suggesting that other factors may contribute to the observed violent crime rates in certain regions. The line plot, which examines the relationship between the average housing price index and the number of violent crimes over time, demonstrates that both variables have experienced fluctuations. Overall, the housing price index has shown an upward trend, while the number of violent crimes exhibits a general decrease, though not consistently. This suggests that, on a broader scale, higher housing prices may be associated with lower violent crime rates. Therefore, the visualizations indicate a negative correlation between the housing price index and violent crime rates in the United States.

#### 5. Relationship between the Housing Price and Salary & Population {#sec-5}

```python
# | echo: false

# housing data set
df3= df2_year.copy()
df3_house=df3.loc[df3["Year"]==2022].reset_index(drop=True)
df3_house["State_long"]=df3_house["State"]

state_mapping = {
    'AK': 'Alaska', 'AL': 'Alabama', 'AR': 'Arkansas', 'AZ': 'Arizona', 'CA': 'California',
    'CO': 'Colorado', 'CT': 'Connecticut', 'DC': 'District of Columbia', 'DE': 'Delaware',
    'FL': 'Florida', 'GA': 'Georgia', 'HI': 'Hawaii', 'IA': 'Iowa', 'ID': 'Idaho',
    'IL': 'Illinois', 'IN': 'Indiana', 'KS': 'Kansas', 'KY': 'Kentucky', 'LA': 'Louisiana',
    'MA': 'Massachusetts', 'MD': 'Maryland', 'ME': 'Maine', 'MI': 'Michigan', 'MN': 'Minnesota',
    'MO': 'Missouri', 'MS': 'Mississippi', 'MT': 'Montana', 'NC': 'North Carolina', 'ND': 'North Dakota',
    'NE': 'Nebraska', 'NH': 'New Hampshire', 'NJ': 'New Jersey', 'NM': 'New Mexico', 'NV': 'Nevada',
    'NY': 'New York', 'OH': 'Ohio', 'OK': 'Oklahoma', 'OR': 'Oregon', 'PA': 'Pennsylvania',
    'RI': 'Rhode Island', 'SC': 'South Carolina', 'SD': 'South Dakota', 'TN': 'Tennessee', 'TX': 'Texas',
    'UT': 'Utah', 'VA': 'Virginia', 'VT': 'Vermont', 'WA': 'Washington', 'WI': 'Wisconsin',
    'WV': 'West Virginia', 'WY': 'Wyoming'
}

df3_house["State_long"]=df3_house["State_long"].map(state_mapping)
df3_house.rename(columns = {'H_idx':'Price_index'}, inplace = True)

# Salaries
df3_salary=pd.read_csv("../data/Salaries_2022.csv")
df3_salary.columns=["State_long","Average_Salary","Median_Salary"]

# Population
df3_population=pd.read_csv("../data/Population_2022.csv")
df3_population.columns=["State_long","FIPS_Code","Population"]
```

```python
# | echo: false

from functools import reduce
df3_list=[df3_house,df3_population,df3_salary]
df3_use = reduce(lambda  left,right: pd.merge(left,right,on=['State_long'],how='outer'), df3_list)
df3_use['Population']=df3_use['Population'].str.replace('\W', '', regex=True)
df3_use['Average_Salary']=df3_use['Average_Salary'].str.replace('\W', '', regex=True)
df3_use['Median_Salary']=df3_use['Median_Salary'].str.replace('\W', '', regex=True)
df3_use["Price_index"]=round(df3_use["Price_index"],2)

df3_use['text1']= df3_use["State_long"]+ '<br>' + "Population: "+ df3_use["Population"].astype("str") + '<br>' + "Housing Price Index: "+ df3_use["Price_index"].astype("str")

df3_use['text2']= df3_use["State_long"]+ '<br>' + "Average Salary: "+ df3_use["Population"].astype("str") + '<br>' + "Housing Price Index: "+ df3_use["Price_index"].astype("str")
```

In conclusion, the average housing prices in the US have shown a general increasing trend, except for a decline from 2007 to 2012. States such as MA, NY, and DC have notably higher average housing prices, while WV has lower average housing prices. Seasonality does not seem to have a significant effect on US housing prices. Furthermore, there is a positive correlation between the housing prices and the average salaries of a state, but the correlation between the housing price index and its population appears to be weak. Additionally, areas with high crime rates tend to experience lower housing prices, with violent crimes having a more significant impact than property crimes. 


It is important to note that while average housing prices may be increasing, this trend may not necessarily be sustainable in the long run.
